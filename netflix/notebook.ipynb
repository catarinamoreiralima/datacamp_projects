```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
netflix_df.head(10)
```

We can see the mode (most frequent) using .mode(). It returns a series with all values that are the mode of the dataframe. 

```python
movies_90s = netflix_df[
    (netflix_df["release_year"] >= 1990) &
    (netflix_df["release_year"] <= 1999)
]
duration = movies_90s["duration"].mode().iloc[0]
duration
```

Using a condition with a dataframe returns a boolean array with every instance that matches the condition. You can then just sum the array (0 and 1) and it will return the ammount of entries that fit that.

```python
short_movie_count = (
    (movies_90s["type"] == "Movie") &
    (movies_90s["genre"] == "Action") &
    (movies_90s["duration"] < 90)
).sum()

print(short_movie_count)
```
